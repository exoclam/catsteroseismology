```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import corner

import matplotlib.pylab as pylab
params = {'legend.fontsize': 'large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large'}
pylab.rcParams.update(params)
```

```python
cats = pd.read_csv('cats.csv')
cats
```

```python
cats['How old is your cat?'].astype('float64').dtypes
cats['If you know, how much does your cat weigh in pounds?'].astype('float64').dtypes
cats['Subjectively, how large is your cat?'].astype('float64').dtypes
cats['How cuddly is your cat?'].astype('float64').dtypes
cats['How aggressive is your cat?'].astype('float64').dtypes
cats['How intelligent is your cat?'].astype('float64').dtypes
cats['How hyper is your cat?'].astype('float64').dtypes
cats['When you pet your cat, what percentage of the time will they start to purr?'].astype('float64').dtypes
cats['How loudly does your cat purr?'].astype('float64').dtypes

cats_sub = cats.rename(columns={"How old is your cat?": "age",
                    "If you know, how much does your cat weigh in pounds?": "weight", 
                     "Subjectively, how large is your cat?": "length",
                    "How cuddly is your cat?": "cuddliness",
                    "How aggressive is your cat?": "aggression",
                    "How intelligent is your cat?": "intelligence",
                    "How hyper is your cat?": "activity level",
                    "When you pet your cat, what percentage of the time will they start to purr?": "purr susceptibility",
                    "How loudly does your cat purr?": "purr volume"})
```

```python
cats_sub = cats_sub.dropna()
```

```python
figure = corner.corner(cats_sub[['age', 'weight', 'length', 'cuddliness', 'aggression', 'intelligence',
                       'activity level', 'purr susceptibility', 'purr volume']])

figure.savefig('corner-cats.pdf', format='pdf')
```
